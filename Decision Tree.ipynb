## Q1. Use Decision Tree to build a model for classification of Iris Dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

**Loading the dataset**

```python
data = load_iris()
```

```python
X = data.data
y = data.target
```

```python
frame = pd.DataFrame(data=X, columns=data.feature_names)

frame.head()
```

**Splitting the dataset into training and testing sample.**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

**Splitting the training sample into training and validation samples.**

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.1, random_state=100)
```

**Training the model.**

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth=3)

model.fit(X_train, y_train)

fig = plt.figure(figsize=(10, 8))
decision_tree = tree.plot_tree(model, filled=True, class_names=data.target_names, feature_names=data.feature_names)
```

**Using validation data to check the accuracy of the model.**

```python
y_pred = model.predict(X_valid)

print(f'Accuracy Score: {accuracy_score(y_valid, y_pred)}')
```

**Predicting the class for the test sample.**

```python
y_pred = model.predict(X_test)

print(f'Accuracy Score: {accuracy_score(y_test, y_pred)}')
```

## Q2. Use Decision Tree to build a model for classification of Breast Cancer Dataset.

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()

X = data.data
y = data.target
```

```python
frame = pd.DataFrame(data=X, columns=data.feature_names)

frame.head()
```

**Splitting the dataset into training and testing sample.**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

**Splitting the training sample for training and validation purpose.**

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.1, random_state=100)
```

**Training the model.**

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth=6)

model.fit(X_train, y_train)

fig = plt.figure(figsize=(100, 30))
decision_tree = tree.plot_tree(model, filled=True, class_names=data.target_names, feature_names=data.feature_names)
```

**Using validation data to check the accuracy of the model.**

```python
y_pred = model.predict(X_valid)

print(f'Accuracy Score: {accuracy_score(y_valid, y_pred)}')
```

**Predicting the class for the test sample.**

```python
y_pred = model.predict(X_test)

print(f'Accuracy Score: {accuracy_score(y_test, y_pred)}')
```
